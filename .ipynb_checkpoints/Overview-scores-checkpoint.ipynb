# Overview Scores

After applying one or multiple algorithms to one or multiple datasets an overview graph can be generated.\
This graph displays the f1-macro scores for each model on each dataset.\
The graph can be saved onder a directory by specifying the `path` argument of the `plot_overview` function.
The path to the most recent directory of outputs is saved under the cache and can be loaded to look at results.

```python
from src.visuals import get_scores, plot_overview

with open("cache/.dir_latest_run", "r") as f:
    path_run = f.read()

if "multirun" in path_run:
    plot_overview(get_scores(path_run), path=path_run)
```

It's also possible to select a path from a previous run. If the `path` argument of plot_overview is not specified no graph will be saved.

```python
base_path = 'multirun'
date = "2023-01-24"
hm = "15-54-04"
path_run = f"{base_path}/{date}/{hm}"

plot_overview(get_scores(path_run))
```
